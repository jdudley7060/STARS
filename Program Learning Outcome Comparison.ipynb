```python
#make sure working directory is also where the file is and that the paths being read in below match your file name
csv_to_check = process_csv("PLOs for certificates and majors.xlsx - Sheet1.csv")
csv_keywords = process_csv("Sample keywords to help identify sustainability challenges - Sheet1.csv")

#saves the header of the PLO's
header_check = csv_to_check[0]

#saves the header of the keywords
header_keywords = csv_keywords[0]

check_data = csv_to_check[1:] # The entire CSV data set besides the headers
keywords_data = csv_keywords[1:]
```

```python
#first turn the orignial data into a list of dicts 

#intialize an empty list
final_list = [] 

#intialize an empty dict to store the courses and their learning outcomes that fit the bill
class_info_i = {}

#begin looping through the loaded in CSV file with the PLO's
for idx in range(len(check_data)):

    prev_val_degree = check_data[idx][1]
    
    #intialize an empty table for each course 
    plo_table_i = []
    
    #loops through the columns to make sure we are in the right one
    for item in range(len(header_check)):
        
        value = check_data[idx][item]
        
        #if the column is PLO
        if header_check[item] == "PLO":
            
            #add the learning objectives to its lists
            if check_data[idx][1] not in class_info_i:
                class_info_i[check_data[idx][1]] = [value]
                continue
            
            elif check_data[idx][1] in class_info_i:
                
                class_info_i[check_data[idx][1]].append(value)
    
    
class_info_i
```

```python
sus_dict = copy.deepcopy(class_info_i)

#create empty list for each major  putting vals in
for key,value in sus_dict.items():
    sus_dict[key] = []

#this first loop looks to see if any of the orginal PLO's match any of the keywords and then saves the PLO if there is a mathc
for key, value in class_info_i.items():
    
    #loops through the sustianability keywords
    for i in range(len(keywords_data)):
        key_i = keywords_data[i][0] 
        
        #loops throughn the list of PLO's nestled withinn the PLO dictionary
        for idx in range(len(value)):
        
        #if the the key word is conatinaed in any of the PLO's (case non-sensitive) then it adds it to thee dictionary
            if key_i in value[idx].lower():
                sus_dict[key].append(value[idx])
                
what_trigger = copy.deepcopy(class_info_i)

#create empty list for putting vals in
for key,value in what_trigger.items():
    what_trigger[key] = []

#this loop does the samme as the above one but instead of svaing the PLO it saves the word that triggered it
for key, value in class_info_i.items():
    
    for i in range(len(keywords_data)):
        key_i = keywords_data[i][0] 
        
        for idx in range(len(value)):
        
            if key_i in value[idx].lower():
                what_trigger[key].append(key_i)
what_trigger
sus_dict

#this here just saves them to seperate CSV files that can then be looked at and compared with ease
a_file = open("Course PLO.csv", "w")
b_file = open("Word Trigger.csv", "w")

writer = csv.writer(a_file)
for key, value in sus_dict.items():
    writer.writerow([key, value])
    
writer = csv.writer(b_file)
    
for key, value in what_trigger.items():
    writer.writerow([key, value])
```
